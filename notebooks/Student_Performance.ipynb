### Splitting and training


```python
X = df.drop(["Pass_Fail", "G3"], axis=1)
y = df["G3"]
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler


processor = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_col),
        ('num', StandardScaler(), num_cols)
    ]
)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.3
)
```

```python
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LassoCV, LinearRegression, RidgeCV
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor


models = {
    "LR": LinearRegression(),
    "Lasso": LassoCV(cv=5, random_state=42),
    "Ridge": RidgeCV(cv=5),
    "DecisionTree": DecisionTreeRegressor(max_depth=8, random_state=42),
    "RandomForest": RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        n_jobs=-1
    ),
    "XGBoost": XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    random_state=42,
    n_jobs=-1
    ),

    "LightGBM": LGBMRegressor(
        n_estimators=300,
        learning_rate=0.05,
        random_state=42,
        n_jobs=-1
    )
}
```

```python
param_grids = {

    "LR": {},

    "Lasso": {
        "model__alphas": [[0.0001, 0.001, 0.01, 0.1, 1, 10]]
    },

    "Ridge": {
        "model__alphas": [[0.01, 0.1, 1, 10, 100]]
    },

    "DecisionTree": {
        "model__max_depth": [3, 5, 8, 10, 15, None],
        "model__min_samples_split": [2, 5, 10, 20],
        "model__min_samples_leaf": [1, 2, 4, 8],
        "model__criterion": ["squared_error", "friedman_mse"]
    },

    "RandomForest": {
        "model__n_estimators": [100, 200, 300, 500],
        "model__max_depth": [None, 10, 20, 30],
        "model__min_samples_split": [2, 5, 10],
        "model__min_samples_leaf": [1, 2, 4],
        "model__max_features": ["sqrt", "log2", None]
    },

    "XGBoost": {
        "model__n_estimators": [100, 200, 300, 500],
        "model__learning_rate": [0.01, 0.03, 0.05, 0.1],
        "model__max_depth": [3, 5, 7, 10],
        "model__subsample": [0.6, 0.8, 1.0],
        "model__colsample_bytree": [0.6, 0.8, 1.0],
        "model__gamma": [0, 0.1, 0.3]
    },

    "LightGBM": {
        "model__n_estimators": [100, 200, 300, 500],
        "model__learning_rate": [0.01, 0.03, 0.05, 0.1],
        "model__max_depth": [-1, 5, 10, 20],
        "model__num_leaves": [15, 31, 63, 127],
        "model__subsample": [0.6, 0.8, 1.0],
        "model__colsample_bytree": [0.6, 0.8, 1.0]
    }

}
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error,
    root_mean_squared_error,
)

# ----------------------------
# Hyperparameter Space
# ----------------------------

param_grids = {

    "LR": {},

    "Lasso": {},

    "Ridge": {},

    "DecisionTree": {
        "model__criterion": ["squared_error", "friedman_mse"],
        "model__max_depth": [None, 5, 10, 15, 20],
        "model__min_samples_split": [2, 5, 10, 20],
        "model__min_samples_leaf": [1, 2, 4, 8]
    },

    "RandomForest": {
        "model__n_estimators": [100, 200, 300, 500],
        "model__max_depth": [None, 10, 20, 30],
        "model__min_samples_split": [2, 5, 10],
        "model__min_samples_leaf": [1, 2, 4],
        "model__max_features": ["sqrt", "log2", None]
    },

    "XGBoost": {
        "model__n_estimators": [100, 200, 300, 500],
        "model__learning_rate": [0.01, 0.03, 0.05, 0.1],
        "model__max_depth": [3, 5, 7, 10],
        "model__subsample": [0.6, 0.8, 1.0],
        "model__colsample_bytree": [0.6, 0.8, 1.0],
        "model__gamma": [0, 0.1, 0.3]
    },

    "LightGBM": {
        "model__n_estimators": [100, 200, 300, 500],
        "model__learning_rate": [0.01, 0.03, 0.05, 0.1],
        "model__max_depth": [-1, 5, 10, 20],
        "model__num_leaves": [15, 31, 63, 127],
        "model__subsample": [0.6, 0.8, 1.0],
        "model__colsample_bytree": [0.6, 0.8, 1.0]
    }

}

# ----------------------------
# Model Tuning
# ----------------------------

results = []
best_models = {}

for name, model in models.items():

    print("=" * 60)
    print(f"Tuning {name}")
    print("=" * 60)

    pipeline = Pipeline(
        steps=[
            ("processor", processor),
            ("model", model)
        ]
    )

    # Models without hyperparameters
    if len(param_grids[name]) == 0:

        pipeline.fit(X_train, y_train)

        best_model = pipeline

        cv_score = cross_val_score(
            pipeline,
            X_train,
            y_train,
            cv=5,
            scoring="r2",
            n_jobs=-1
        )

        best_params = "Default"

    else:

        search = RandomizedSearchCV(
            estimator=pipeline,
            param_distributions=param_grids[name],
            n_iter=20,
            cv=5,
            scoring="r2",
            random_state=42,
            n_jobs=-1,
            verbose=1
        )

        search.fit(X_train, y_train)

        best_model = search.best_estimator_

        best_params = search.best_params_

        cv_score = cross_val_score(
            best_model,
            X_train,
            y_train,
            cv=5,
            scoring="r2",
            n_jobs=-1
        )

    # Save model
    best_models[name] = best_model

    # Predictions
    y_pred = best_model.predict(X_test)

    # Metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = root_mean_squared_error(y_test, y_pred)

    results.append({

        "Model": name,
        "R2": round(r2, 4),
        "MAE": round(mae, 4),
        "RMSE": round(rmse, 4),
        "MSE": round(mse, 4),
        "CV Mean": round(cv_score.mean(), 4),
        "CV Std": round(cv_score.std(), 4),

    })
```

```python

# ----------------------------
# Results
# ----------------------------

results = pd.DataFrame(results)

results = results.sort_values(
    by="R2",
    ascending=False
).reset_index(drop=True)

print(results)
```

```python

# ----------------------------
# Best Model
# ----------------------------

best_name = results.iloc[0]["Model"]

print("\nBest Model :", best_name)

best_model = best_models[best_name]
```

```python
best_model = best_models["XGBoost"]

import joblib

joblib.dump(best_model, "../model/best_model.pkl")
```
